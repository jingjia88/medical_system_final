```python
from folium.plugins import HeatMap
data = []
for i in range(len(f)):
    data.append([f[i]['coordinates']['lat'], f[i]['coordinates']['lng'],f[i]['populartimes'][0]['data'][19]])
下面location = 地圖的中心點，以下以台北市為例，或許可以取上面p1,p2的中心位置
hmap = folium.Map(location=[25.030724, 121.520076], tiles='stamentoner',control_scale=True, zoom_start=13)
hmap.add_child(HeatMap(data, radius=10, gradient={.25: 'blue', .5: 'lime', 0.75: 'red'}))
hmap
# 將圖存入test1.html
# file_path = r"test1.html"
# hmap.save(file_path)  # 
```

```python
import folium.plugins as plugins

data_move = []
m = folium.Map([(p1[0]), (p1[1]+p2[1])/2], tiles='stamentoner', zoom_start=13)


for hour in range(24):
    tmp = []
    for i in range(len(f)):
        tmp.append([f[i]['coordinates']['lat'], f[i]['coordinates']['lng'],f[i]['populartimes'][0]['data'][hour]])
    data_move.append(tmp)
#print(data_move[0])
print(len(data_move))
print(data_move[0][0])
print(data_move[1][0])

plugins.HeatMapWithTime(data_move,radius=5,auto_play=True,position='bottomright',name="cluster",max_opacity=0.7).add_to(m)
m
# hm = plugins.HeatMapWithTime(data_move, auto_play=True,radius=15)
# hm.add_to(m)
# hm
# file_path = r"test1.html"
# hm.save(file_path)
```
